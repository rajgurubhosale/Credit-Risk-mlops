```python
SPECIAL_VALUES = [-99999, -88888, -77777]

# special values
special_df = pd.DataFrame({
    'feature': df.columns,
    'special_ratio': df.isin(SPECIAL_VALUES).sum().values / df.shape[0]
})

#zero values
zero_df = (
    (df == 0).sum()
    .div(df.shape[0])
    .reset_index()
    .rename(columns={'index': 'feature', 0: 'zero_ratio'})
)

#Null_values
new_df = (df.isnull().sum() / df.shape[0]).to_frame('null_ratio').reset_index().rename(columns={'index': 'feature'})
```

```python
feature_quality_df = special_df.merge(zero_df,on='feature',how='left').merge(new_df,on='feature',how='left')
feature_quality_df
```

```python
feature_quality_df['max_bad_ratio'] = (
    feature_quality_df[['special_ratio', 'zero_ratio', 'null_ratio']]
    .max(axis=1)
)
feature_quality_df['drop_feature'] = (
    feature_quality_df['max_bad_ratio'] > 0.95
)
```

```python
temp = feature_quality_df[feature_quality_df['drop_feature']==True]
drop_features = temp['feature'].values.tolist()
```
